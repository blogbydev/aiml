The <class 'tensorflow.python.data.ops.prefetch_op._PrefetchDataset'> message indicates that you're working with a TensorFlow dataset and specifically referring to a dataset that's being prefetched for better performance during training.

What is _PrefetchDataset?
In TensorFlow, the PrefetchDataset is an optimization technique where data is preloaded (or prefetched) onto the CPU or GPU while the model is training, reducing the waiting time for data to be available when needed. This is especially useful when working with large datasets or slow data sources, like reading from disk.

This _PrefetchDataset class is part of TensorFlow’s tf.data API, and it's used under the hood when you use the .prefetch() method in a dataset pipeline.

```python
file_paths = np.array(data.file_paths)
labels = np.array([label.numpy()[0] for _, label in data])
```

```python

# Loading images one by one with batch_size=1 and not shuffling them (shuffle=False) is particularly useful
# if you want to inspect or manipulate the data before splitting it into training, validation, and test sets.
# This allows you to ensure that the data is split accurately and consistently without any unintended randomization.

# Extract file paths and labels
# file_paths = np.array(data.file_paths)
# labels = np.array([label.numpy()[0] for _, label in data]) # iterate over images with their corresponding labels
# label.numpy()[0]: Extracts the scalar value from the NumPy array.
# Since batch_size=1, each label is in a batch of size 1,
# so label.numpy() will return an array of shape (1,).
# Accessing the first (and only) element with [0] gives the scalar value (either 0 or 1).

# Split into train, validation, and test sets (60% train, 20% validation, 20% test)

train_paths, test_paths, train_labels, test_labels = train_test_split(file_paths, labels, test_size=0.4, random_state=42)
val_paths, test_paths, val_labels, test_labels = train_test_split(test_paths, test_labels, test_size=0.5, random_state=42)

# Verify the splits
print(f"Training set size: {len(train_paths)}")
print(f"Validation set size: {len(val_paths)}")
print(f"Test set size: {len(test_paths)}")
```

```python
def create_dataset(file_paths, labels):
    # Load and preprocess the images
    def load_and_preprocess_image(file_path, label):
        img = tf.io.read_file(file_path) # Reads the image file from the specified file_path as a raw byte string.
        img = tf.image.decode_jpeg(img, channels=3) # Decodes the raw byte string into a tensor representing the image. Since the images are JPEG files, decode_jpeg is used. The channels=3 parameter ensures the image is decoded with 3 color channels (RGB), even if it's grayscale.
        img = tf.image.resize(img, [320, 240]) #Resizes the image to the specified dimensions of 320x240 pixels.
        img = img / 255.0  # Normalize to [0, 1] range
        return img, label

    # Create a TensorFlow dataset
    dataset = tf.data.Dataset.from_tensor_slices((file_paths, labels)) # it creates a dataset where each element is a pair consisting of a file path and its corresponding label.
    dataset = dataset.map(load_and_preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
    # TensorFlow handles the parallel processing of elements. This means multiple images can be processed at the same time, speeding up the data preparation phase.
    return dataset

# Create datasets
train_ds = create_dataset(train_paths, train_labels)
val_ds = create_dataset(val_paths, val_labels)
test_ds = create_dataset(test_paths, test_labels)
```

```python
import matplotlib.pyplot as plt

def visualize_dataset(dataset, title, num_samples=5):
    plt.figure(figsize=(15, 10))
    total_images = num_samples * 5  # Maximum number of images to visualize
    count = 0

    for images, labels in dataset.take(num_samples):
        for j in range(images.shape[0]):
            if count >= total_images:
                break
            plt.subplot(num_samples, 5, count + 1)
            plt.imshow(images[j].numpy())
            plt.title(f"Label: {'Daisy' if labels[j].numpy() == 0 else 'Non-Daisy'}")
            plt.axis('off')
            count += 1

        if count >= total_images:
            break

    plt.suptitle(title, fontsize=16)
    plt.show()

# Example usage with batch size = 32
visualize_dataset(train_ds.batch(32), 'Training Dataset Samples')
visualize_dataset(val_ds.batch(32), 'Validation Dataset Samples')
visualize_dataset(test_ds.batch(32), 'Test Dataset Samples')
```

```python
# Configure datasets for performance
BATCH_SIZE = 32
#number of samples /batch size = number of batches per epoch
AUTOTUNE = tf.data.experimental.AUTOTUNE  # setting that allows TensorFlow to automatically tune the performance of your data pipeline.
# It dynamically adjusts the buffer size used in prefetching based on the available system resources (like CPU and memory).
train_ds = train_ds.shuffle(buffer_size=len(train_paths)).batch(BATCH_SIZE).prefetch(buffer_size=AUTOTUNE) # The buffer size determines how many elements from the dataset TensorFlow will load into memory at once for shuffling.
val_ds = val_ds.batch(BATCH_SIZE).prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.batch(BATCH_SIZE).prefetch(buffer_size=AUTOTUNE)
# Prefetching allows the data pipeline to prepare the next batch of data while the current batch is being processed by the model.
# This overlap of computation and data loading helps to keep the GPU/CPU busy, reducing idle time and improving training speed.
```

Batch Size:
Typical Values:

16, 32, 64: These are the most common starting points.
128, 256: May be used if you have a large dataset and sufficient memory.
Considerations:

Smaller Batch Size: May lead to better generalization but can be slower to train.
Larger Batch Size: Often faster training but may require a lower learning rate and more careful tuning to avoid poor generalization.
Epochs:
Typical Values:

10-50 epochs: For simpler problems or when you expect quick convergence.
50-200 epochs: For more complex problems or when the dataset is large.
Considerations:

Early Stopping: Implement early stopping based on validation performance to avoid overfitting.
Monitor Loss: Keep an eye on training and validation loss; stop if you notice validation loss increasing while training loss decreases.

Batch Size:
Small Datasets: For small datasets, you can use a larger batch size (e.g., 32, 64, or even 128) to speed up training, as it will have more reliable gradient estimates.
Large Datasets: For large datasets, smaller batch sizes (e.g., 16 or 32) may work better, as they introduce more noise into the gradient, which can help in avoiding local minima and improving generalization.
Memory Constraints: The batch size is often limited by the GPU memory. Larger models or higher resolution images may require smaller batch sizes due to memory constraints.
Common Practice: Common batch sizes range from 16 to 128, with 32 or 64 being typical starting points.

```python
from tensorflow.keras import layers, models

# Build the ANN model
ann_model = models.Sequential([
    layers.Flatten(input_shape=(320, 240, 3)),  # Flatten the 2D image to a 1D vector
    layers.Dense(128, activation='relu'),      # Hidden layer with 128 neurons
    layers.Dropout(0.5),                       # Dropout for regularization
    layers.Dense(64, activation='relu'),       # Hidden layer with 64 neurons
    layers.Dense(1, activation='sigmoid')      # Output layer for binary classification
])

# dropout it will create mask vector - percentage for weights to be zero


# Compile the ANN model
ann_model.compile(optimizer='adam',              # Adam optimizer
                  loss='binary_crossentropy',    # Binary crossentropy loss
                  metrics=['accuracy'])          # Accuracy as the evaluation metric
```

```python
 # Train the ANN model
ann_history = ann_model.fit(
    train_ds,                  # Training dataset
    validation_data=val_ds,    # Validation dataset
    epochs=10                  # Number of epochs
)
```

Loss Decrease: The model shows a steady decrease in training loss, indicating it is learning the underlying patterns in the data.
Stabilizing Validation Metrics: Validation loss and accuracy stabilize after a few epochs, suggesting the model is neither overfitting nor underfitting significantly.
Accuracy Increase: Both training and validation accuracy improve over epochs, but the improvement slows down towards the end.

Training Improvement: The training accuracy improves consistently, and the training loss decreases steadily, indicating that the model is learning effectively from the training data.

Validation Performance: Validation accuracy shows fluctuations, peaking around Epoch 4 and then gradually declining. The increasing validation loss in the later epochs suggests that the model may be overfitting the training data.

```python
# Plot training & validation accuracy values
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(ann_history.history['accuracy'])
plt.plot(ann_history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Training Accuracy', 'Validation Accuracy'])

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(ann_history.history['loss'])
plt.plot(ann_history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Training Loss', 'Validation Loss'])

plt.show()
```

```python
# Evaluate the ANN model on the test dataset
ann_test_loss, ann_test_acc = ann_model.evaluate(test_ds)
print(f"ANN Test accuracy: {ann_test_acc:.2f}")
```

Why Use SciKeras?
SciKeras bridges the gap between Keras (a deep learning library) and Scikit-learn (a machine learning library). It enables you to use Keras models as if they were Scikit-learn models, providing seamless integration for tasks like grid search, cross-validation, and pipeline construction.

```python
!pip install scikeras

import numpy as np
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score  # Import accuracy_score

# Define the ANN model
def create_ann_model(neurons=128, dropout_rate=0.5, learning_rate=0.001):
    model = models.Sequential([
        layers.Flatten(input_shape=(320, 240, 3)),
        layers.Dense(neurons, activation='relu'),
        layers.Dropout(dropout_rate),
        layers.Dense(neurons // 2, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# Wrap the model with KerasClassifier
model = KerasClassifier(model=create_ann_model, verbose=0)
# KerasClassifier: Wraps the Keras model so it can be used with Scikit-learn tools like GridSearchCV.
# verbose=0: Suppresses training logs for clarity.
# Simplified hyperparameter grid with fewer values
param_grid = {
    'model__neurons': [64, 128],
    'model__dropout_rate': [0.3, 0.5],
    'model__learning_rate': [0.001],
    'batch_size': [16],
    'epochs': [5]  # Reduced epochs for faster training
}

# Convert TensorFlow Dataset to NumPy arrays
def dataset_to_numpy(dataset):
    X, y = [], []
    for images, labels in dataset:
        X.extend(images.numpy())
        y.extend(labels.numpy())
    return np.array(X), np.array(y)

X_train, y_train = dataset_to_numpy(train_ds)

# Grid search for best parameters
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, verbose=2, n_jobs=-1)
grid_result = grid.fit(X_train, y_train)

# Display best parameters and test accuracy
print(f"Best Parameters: {grid_result.best_params_}")
print(f"Best Validation Accuracy: {grid_result.best_score_:.2f}")

# Evaluate on the test set
X_test, y_test = dataset_to_numpy(test_ds)

# Get predictions using predict_proba to get probabilities for class 1
y_pred_proba = grid_result.best_estimator_.predict_proba(X_test)[:, 1]


# Convert probabilities to class labels (0 or 1) using a threshold of 0.5
y_pred = (y_pred_proba > 0.5).astype(int)

# Calculate and print test accuracy
test_acc = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {test_acc:.2f}")

# Visualize and predict on a single image
for images, labels in test_ds.take(1):
    first_image, first_label = images[0].numpy(), labels[0].numpy()
    break

plt.imshow(first_image)
plt.title(f"Actual Label: {'Daisy' if first_label == 0 else 'Non-Daisy'}")
plt.show()

first_image = first_image.reshape(1, *first_image.shape)
# Get prediction probability
# For a single image, the output might look like: [[0.3, 0.7]]
# for a batch of images [0.3,0.7] for every image
prediction_proba = grid_result.best_estimator_.predict_proba(first_image)
# Check if prediction_proba is 2D, if not, use the first value
# This ensures that prediction_proba is a single value
prediction_proba = prediction_proba[0] if prediction_proba.ndim == 2 else prediction_proba[0]
predicted_label = "Non-Daisy" if prediction_proba > 0.5 else "Daisy"

print(f"Predicted Label: {predicted_label}")
```
