```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)  # 1000 data points from a normal distribution

# Plot KDE
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Using seaborn's kdeplot function to plot the KDE
sns.kdeplot(data, shade=True, color="blue")

# Adding titles and labels
plt.title("Kernel Density Estimate of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Density")

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)  # 1000 data points from a normal distribution

# Plot a histogram
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Using seaborn's displot to plot a histogram
sns.histplot(data, kde=True, bins=30, color='blue')

# Adding titles and labels
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```
