```python
import numpy as np
import matplotlib.pyplot as plt

years = np.array(['2012', '2013', '2014', '2015'])

sales_africa = np.array([127187.27, 144480.70, 229068.79, 283036.44])

sales_USCA = np.array([492756.60, 486629.30, 627634.98, 757108.13])

sales_LATAM = np.array([385098.15, 464733.29, 608140.77, 706632.93])

sales_Asia_Pacific = np.array([713658.22, 863983.97, 1092231.65, 1372784.40])

sales_Europe = np.array([540750.63, 717611.40, 848670.24, 1180303.95])
```

```python
plt.title("Year wise sales of regions", fontdict={"fontsize": 20, "fontweight": 8, "color": "Blue"})
plt.xlabel(xlabel="Years", fontdict={"fontsize":12, "fontweight":8, "color": "brown"})
plt.ylabel(ylabel="Sales", fontdict={"fontsize":12, "fontweight":8, "color": "brown"})
yticks = range(100000, 1500000, 100000)
ytickslables = [f'{item//1000}k' for item in yticks]
plt.yticks(ticks=yticks, labels=ytickslables)
# plt.xticks(ticks=range(2012,2016), labels=range(2012,2016))

plt.plot(years, sales_africa, color="orange", dashes=[2,2,2,2], marker="o", label="Africa")
plt.plot(years, sales_USCA, color="red", dashes=[2,2,2,2], marker="o", label="USCA")
plt.plot(years, sales_LATAM, color="blue", dashes=[2,2,10,2], marker="o", label="LATAM")
plt.plot(years, sales_Asia_Pacific, color="brown", dashes=[2,2,10,2], marker="o", label="APAC")
plt.plot(years, sales_Europe, color="green", dashes=[2,2,10,2], marker="o", label="Europe")
plt.legend()
plt.show()
```
