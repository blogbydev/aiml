```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('census+data.xlsx', sheet_name='C-08', skiprows=6)
df = df[[
    'Unnamed: 0',
    'Unnamed: 1',
    'Unnamed: 2',
    'Unnamed: 3',
    'Unnamed: 4',
    'Unnamed: 5',
    'Unnamed: 6',
    'Unnamed: 7',
    'Unnamed: 8',
    'Unnamed: 9',
    'Unnamed: 10',
    'Unnamed: 11',
    'Unnamed: 12',
    'Unnamed: 13',
    'Unnamed: 14'
]]
df.rename(columns={
    'Unnamed: 0': 'Table Name',
    'Unnamed: 1': 'State Code',
    'Unnamed: 2': 'Distt. Code',
    'Unnamed: 3': 'Area Name',
    'Unnamed: 4': 'Total/Rural/Urban',
    'Unnamed: 5': 'Age-group',
    'Unnamed: 6': 'Total Persons',
    'Unnamed: 7': 'Total Males',
    'Unnamed: 8': 'Total Females',
    'Unnamed: 9': 'Illiterate Persons',
    'Unnamed: 10': 'Illiterate Males',
    'Unnamed: 11': 'Illiterate Females',
    'Unnamed: 12': 'Literate Persons',
    'Unnamed: 13': 'Literate Males',
    'Unnamed: 14': 'Literate Females'
}, inplace=True)
df.head()
```

```python
row_condition_1 = df['Age-group'] == '20-24'
row_condition_2 = df['Area Name'] == 'INDIA'
row_condition_3 = df['Total/Rural/Urban'] == 'Total'
foo = df[row_condition_1 & row_condition_2 & row_condition_3]
foo['Illiterate Females']/foo['Total Females']
```

```python
df['Literacy Rate'] = df['Literate Persons']/df['Total Persons']
row_condition_2 = df['Area Name'] == 'INDIA'
row_condition_3 = df['Total/Rural/Urban'] == 'Total'
data = df[row_condition_2 & row_condition_3]
```

```python
sns.barplot(data=data, x='Age-group', y='Literacy Rate')
plt.xticks(rotation=90)
plt.show()
```

```python
df['female literacy rate'] = df['Literate Females']/df['Total Females']
row_condition_1 = df['Total/Rural/Urban'] == 'Total'
row_condition_2 = df['Age-group'] == 'All ages'
row_condition_3 = df['Area Name'] != 'INDIA'
data = df[row_condition_1 & row_condition_2 & row_condition_3][['Area Name', 'female literacy rate']]
```

```python
sns.barplot(data=data, x='Area Name', y='female literacy rate')
plt.xticks(rotation=90)
plt.show()
```
