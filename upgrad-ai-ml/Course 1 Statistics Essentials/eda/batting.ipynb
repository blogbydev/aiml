```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('odi-batting.csv')
df.head()
```

```python
df['strike rate'] = (df['Runs']*100)/df['Balls']
```

```python
data = df[df['century']][[
    'Player',
    'strike rate'
]]
data = data.sort_values(by=['strike rate'], ascending=False)
data.head()
```

```python
df['year'] = pd.DatetimeIndex(df['MatchDate']).year
df.head()
```

```python
data = df[[
    'Country',
    'century',
    'year'
]]
data = data[data['century']]
data = data.pivot_table(index='Country', columns=['year'], values='century', aggfunc='count')
plt.figure(figsize=(50,5))
sns.heatmap(data, cmap="Greens", annot=True)
plt.show()
```
